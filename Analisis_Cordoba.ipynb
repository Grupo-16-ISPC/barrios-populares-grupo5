#Procesamiento y análisis de datos

Procedemos a filtar los datos para poder analizar en detalle, en este caso, la provincia de Córdoba solamente. 

```python
#Verificamos los datos a filtrar
print(data['provincia'].unique())
```

```python
#Esta funcion crea una serie booleana que es TRUE para las filas donde el valor provincia
#coincide con provincia_deseada. Al pasar esta Serie booleana dentro del DataFrame _data_
#se seleccionarán solo las filas que cumplen esa condición, es decir, las que pertenecen a la provincia deseada.
provincia_deseada = 'Córdoba'  

data_filtrado = data[data['provincia'] == provincia_deseada]
```

```python
print(data_filtrado) #Visualizamos que este correcto el filtrado
```

Agrupamos las columnas y realizamos un conteo, luego imprimimos para visualizar la cantidad de barrios por departamento. 

```python
conteo_barrios = data_filtrado.groupby('departamento')['nombre_barrio'].value_counts()
pd.set_option('display.max_rows', None)
print(conteo_barrios)
```

Realizamos el conteo de barrios por departamentos y graficamos para poder tener una visualizacionmas clara.

```python
conteo_departamentos = data_filtrado['departamento'].value_counts()


# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
conteo_departamentos.plot(kind='bar')
plt.xlabel('departamento')
plt.ylabel('conteo_departamentos')
plt.title('Cantidad de Barrios por Departamento en Córdoba')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor legibilidad

plt.show()
```
