# ¿Cuáles son los barrios con mayor densidad poblacional de Tucumán? y cuantas familias viven allí

```python
#Creamos una nueva tabla en nuestro dataframe df_Tucuman que refleje la cantidad total de personas que desprende de las columnas 
#'personas_genero_masc','personas_genero_fem' y 'personas_genero_otrx'

df_Tucuman ['personas_total'] = df_Tucuman['personas_genero_masc']+df_Tucuman['personas_genero_fem']+df_Tucuman['personas_genero_otrx']


#Creamos una nueva columna que nos refleje la densidad poblacional por m2 de cada barrio:

df_Tucuman ['densidad_m2'] = df_Tucuman['personas_total'] / df_Tucuman['superficie_m2']
df_Tucuman
```

```python
#Creamos una nueva variable, donde en ella agrupamos el dat 
densidad_barrios = df_Tucuman.groupby(['id_renabap','nombre_barrio']).densidad_m2.sum().sort_values(ascending=False)
densidad_barrios.head(10).plot.bar()
```

```python
familia_barrios_densidad=df_Tucuman.groupby(['id_renabap','nombre_barrio','cantidad_familias_aproximada']).densidad_m2.sum().sort_values(ascending=False)
familia_barrios_densidad.head(10)
```
