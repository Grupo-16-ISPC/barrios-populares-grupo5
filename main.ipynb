```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/pablo/carpetas_escritorio/Técnicatura en ciencias de datos e inteligencia artificial/final_datos/barrios-populares-grupo5/data/data.csv')
```

```python
filtro_castelli = (df['provincia'] == 'Chaco') & (df['localidad'] == 'Juan José Castelli')
df_castelli = df[filtro_castelli]
```

```python
'''
Los datos que tienes son de un tipo de formato llamado Well-Known Text (WKT)
que se utiliza en SIG (Sistemas de Información Geográfica). 
Para graficar estos datos georeferenciados, vamos a utilizar geopandas.
Este código convertirá tu cadena WKT en un objeto geométrico con la función wkt.loads(), 
luego creará un GeoDataFrame con este objeto y finalmente trazará el polígono en un gráfico.
'''
import geopandas as gpd
from shapely import wkt


# Esto es el dato en formato WKT
#wkt_data = 'MULTIPOLYGON (((-60.633038 -25.933384, -60.633848 -25.931762, -60.63349604192133 -25.93161329170304, -60.63387787510913 -25.93094670829688, -60.633098 -25.930576, -60.632662 -25.931298, -60.631758 -25.930899, -60.630932 -25.932499, -60.633038 -25.933384)))'
# Convertir la cadena WKT a un objeto geométrico
geom = wkt.loads(poligono_castelli)

# Crear un GeoDataFrame
gdf = gpd.GeoDataFrame(geometry=[geom])

# Graficar
gdf.plot()
```

```python
# Convertir la columna geometry de tu dataframe a objetos geométricos
#df_castelli['geometry'] = df_castelli['geometry'].apply(wkt.loads)

nombre_barrios = df_castelli.geometry


# Crear un nuevo GeoDataFrame con la columna geometry actualizada
gdf = gpd.GeoDataFrame(geometry=nombre_barrios)

# Graficar los polígonos de los barrios
gdf.plot()

# Personalizar el gráfico
plt.title('Polígonos de Barrios')
plt.xlabel('Longitud')
plt.ylabel('Latitud')

plt.show()
```
